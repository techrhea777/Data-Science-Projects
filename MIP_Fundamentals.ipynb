##Display different channels of a RGB image

```python
import cv2
from matplotlib import pyplot as plt

# Read an RGB image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your RGB image
rgb_image = cv2.imread(image_path)


# Check if the image is loaded successfully
if rgb_image is None:
    print("Error: Unable to load the RGB image.")
else:
    # Split the RGB image into its channels
    blue_channel, green_channel, red_channel = cv2.split(rgb_image)

    # Set a larger figure size
    plt.figure(figsize=(12, 4))

    # Display the original RGB image
    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
    plt.title('Original RGB Image')

    # Display the Red channel
    plt.subplot(1, 4, 2)
    plt.imshow(red_channel, cmap='gray')
    plt.title('Red Channel')

    # Display the Green channel
    plt.subplot(1, 4, 3)
    plt.imshow(green_channel, cmap='gray')
    plt.title('Green Channel')

    # Display the Blue channel
    plt.subplot(1, 4, 4)
    plt.imshow(blue_channel, cmap='gray')
    plt.title('Blue Channel')

    plt.show()
```

##Readng a RGB image and converting it to Grayscale image

```python
import cv2
from matplotlib import pyplot as plt

# Read an RGB image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your RGB image
rgb_image = cv2.imread(image_path)

# Check if the image is loaded successfully
if rgb_image is None:
    print("Error: Unable to load the RGB image.")
else:
    # Convert RGB image to grayscale
    grayscale_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)

    # Display the original RGB image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
    plt.title('Original RGB Image')

    # Display the converted grayscale image
    plt.subplot(1, 2, 2)
    plt.imshow(grayscale_image, cmap='gray')
    plt.title('Grayscale Image')

    plt.show()
```

##Histogram Equalisation

```python
import cv2
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_8.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and enhanced images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')

plt.show()
```

## Contrast Stretching

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply contrast stretching
min_intensity = 50
max_intensity = 200
stretched_image = np.clip((image - min_intensity) * 255 / (max_intensity - min_intensity), 0, 255).astype(np.uint8)

# Display the original and enhanced images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast Stretched Image')

plt.show()
```

##Sharpening the Image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path)

# Create a sharpening kernel
kernel = np.array([[-1, -1, -1],
                   [-1,  9, -1],
                   [-1, -1, -1]])

# Apply convolution with the sharpening kernel
sharpened_image = cv2.filter2D(image, -1, kernel)

# Display the original and sharpened images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')

plt.show()
```

##Image segmentation using Binary Thresholding

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply a simple thresholding technique (binary thresholding)
_, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Display the original and segmented images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Segmented Image')

plt.show()
```

##Morphological Operations - Erosion

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a kernel for erosion
kernel = np.ones((5, 5), np.uint8)

# Apply erosion
eroded_image = cv2.erode(image, kernel, iterations=1)

# Display the original and eroded images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(eroded_image, cmap='gray')
plt.title('Eroded Image')

plt.show()
```

##Morphological Operation - Dilation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a kernel for dilation
kernel = np.ones((5, 5), np.uint8)

# Apply dilation
dilated_image = cv2.dilate(image, kernel, iterations=1)

# Display the original and dilated images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(dilated_image, cmap='gray')
plt.title('Dilated Image')

plt.show()
```

##Opening (Erosion followed by Dilation)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a kernel for opening
kernel = np.ones((5, 5), np.uint8)

# Apply opening
opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Display the original and opened images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(opened_image, cmap='gray')
plt.title('Opened Image')

plt.show()
```

##Closing(Dilation followed by Erosion)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read an image using OpenCV
image_path = '/content/drive/MyDrive/test_6.png'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a kernel for closing
kernel = np.ones((5, 5), np.uint8)

# Apply closing
closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# Display the original and closed images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(closed_image, cmap='gray')
plt.title('Closed Image')

plt.show()
```
