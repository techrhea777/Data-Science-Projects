```python
import nibabel as nib
import matplotlib.pyplot as plt

# Replace 'path/to/your/nii/image.nii' with the actual path to your NIfTI image
nii_path = '/content/drive/MyDrive/volume-40.nii'
#nii_path = '/content/drive/My Drive/Colab Notebooks/Lab07/liverNifti/segmentation-40.nii'

# Read the NIfTI image
nii_image = nib.load(nii_path)

# Access the image data
nii_data = nii_image.get_fdata()

# Display a slice of the NIfTI image using matplotlib
plt.imshow(nii_data[:, :, nii_data.shape[2] // 2], cmap='gray')
plt.title('NIfTI Image Slice')
plt.show()
```

Keep in mind that NIfTI images can be 3D or 4D (with multiple volumes, e.g., for time series data), and you can adapt the code accordingly based on the structure of your NIfTI files.

If you are working with 4D data and want to visualize multiple volumes over time, you can modify the code accordingly:

```python
# Display multiple slices of the NIfTI image over time
for i in range(nii_data.shape[2]):
    plt.imshow(nii_data[:, :, i], cmap='gray')
    plt.title(f'NIfTI Image Slice {i + 1}')
    plt.show()
```
