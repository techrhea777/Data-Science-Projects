```python
import pandas as pd
from sklearn.datasets import make_classification
#from sklearn.model_selection import train_test_split
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.feature_selection import SelectKBest, f_classif
#from sklearn.metrics import accuracy_score
```

```python
# Generate a random dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
```

```python
# Convert the array into a DataFrame
columns =[f"Feature_{i}" for i in range(X.shape[1])]
df = pd.DataFrame(X, columns=columns)
X_copy = df
y_copy = pd.DataFrame(y)
df['Target'] = y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = X_copy.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
# Calculate Pearson correlation coefficients between features and target
correlations = X_copy.corrwith(y_copy)
```

```python
# Select features with high absolute correlation with the target
selected_features = correlations.abs().nlargest(10).index
```

```python
final_selected_features = []
for features in selected_features:
    if features in columns:
        final_selected_features.append(features)

final_selected_features
```

```python
X_selected = X_copy[final_selected_features]
```

```python
import seaborn as sns
correlation_matrix = X_selected.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
