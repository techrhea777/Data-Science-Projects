```python
import pydicom
import matplotlib.pyplot as plt

# Replace 'path/to/your/dicom/image.dcm' with the actual path to your DICOM image

dicom_path = '/content/drive/MyDrive/CT000000.dcm'

# Read the DICOM image
dicom_image = pydicom.dcmread(dicom_path)

# Extract pixel data
pixel_array = dicom_image.pixel_array

# Display the DICOM image using matplotlib
plt.imshow(pixel_array, cmap='gray')
plt.title('DICOM Image')
plt.show()
```

Keep in mind that DICOM images can contain various metadata related to medical imaging, such as patient information, acquisition parameters, and more. You can access this information through the dicom_image object.

```python
import pydicom

# Replace 'path/to/your/dicom/image.dcm' with the actual path to your DICOM image
dicom_path = '/content/drive/MyDrive/CT000002.dcm'
# Read the DICOM image
dicom_image = pydicom.dcmread(dicom_path)

# Display metadata information
print("DICOM Metadata:")
for element in dicom_image:
    print(f"{element.name}: {element.value}")

# Access specific metadata attributes
print(f"Patient Name: {dicom_image.PatientName}")
print(f"Study Date: {dicom_image.StudyDate}")
print(f"Modality: {dicom_image.Modality}")
# Add more attributes as needed
```

```python
import os

directory_path = '/content/drive/MyDrive'

# Read all DICOM files in the directory
dicom_images = [pydicom.dcmread(os.path.join(directory_path, file)) for file in os.listdir(directory_path) if file.endswith('.dcm')]

# Access pixel data or other DICOM metadata as needed
# Extract pixel data from each slice
slice_data = [slice.pixel_array for slice in dicom_images]

# Display multiple slices
num_slices = len(slice_data)
fig, axes = plt.subplots(1, num_slices, figsize=(15, 5))

'''
for i in range(num_slices):
    axes[i].imshow(slice_data[i], cmap='gray')
    axes[i].set_title(f'Slice {i + 1}')
#plt.show() '''

for i in range(num_slices):
  plt.imshow(slice_data[i], cmap='gray')
  plt.title(f'Dicom Image Slice {i}')
  plt.show()
```
